```python
import pandas as pd
import numpy as np
import os
from tqdm._tqdm_notebook import tnrange

filepath='D:\\research\\8_GUCD\\POI_MN.csv'

file_pd=pd.read_csv(filepath,header=None)
```

```python
file_pd.columns=['place_id','spatial_id','chain_id','category_ids','latitude','longitude','names','address','city','state','zip-code','cbg','dma_id','cbsa_id','local_timezone','opening_date','closing_date','is_open','is_high_confidence','is_multiuse','closed_hours','ceiling_scores']
```

```python
split_pattern = r'[,-]+'
file_category=file_pd.category_ids.str.split(pat=split_pattern, expand=True)
```

```python
file_pd
```

```python
merged_df = pd.concat([file_pd, file_category],axis=1)
```

```python
merged_df
```

```python
contains_piq =merged_df.eq('piq').any(axis=1)
merged_df=merged_df[contains_piq]
```

```python
file_category
```

```python
row_list=[]
col_list=[]
target ='piq'
for index, row in merged_df.iterrows():
    for col in range(len(row)):
        if row[col] == target:
            row_list.append(index)
            col_list.append(col)
            break
```

```python
loca = [[row_list[i], col_list[i]] for i in range(len(row_list))]
```

```python
merged_df.loc[loca[10][0],loca[10][1]+1]
```

```python
category=[]
for i in range(len(loca)):
    category.append(merged_df.loc[loca[i][0],loca[i][1]+1])
```

```python
merged_df['category']=category
```

```python
merged_df['category'].unique()
```

```python
merged_df
```

```python
result=merged_df[['place_id','spatial_id','chain_id','category_ids','latitude','longitude','names','address','city','state','zip-code','cbg','dma_id','cbsa_id','local_timezone','opening_date','closing_date','is_open','is_high_confidence','is_multiuse','closed_hours','ceiling_scores','category']]
```
